```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

# plotting style
sns.set(style="whitegrid")
```

```python
train_df = pd.read_excel("../data/train(1).xlsx")

train_df.head()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(train_df["price"], bins=50, kde=True)
plt.title("Distribution of House Prices")
plt.xlabel("Price")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(
    x=train_df["sqft_living"],
    y=train_df["price"],
    alpha=0.3
)
plt.title("Price vs Living Area (sqft_living)")
plt.xlabel("Living Area (sqft)")
plt.ylabel("Price")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(
    x=train_df["bedrooms"],
    y=train_df["price"]
)
plt.title("Price Distribution by Number of Bedrooms")
plt.xlabel("Bedrooms")
plt.ylabel("Price")
plt.show()
```

```python
plt.figure(figsize=(8,6))

corr = train_df[
    ["price", "bedrooms", "bathrooms", "sqft_living", "lat", "long"]
].corr()

sns.heatmap(
    corr,
    annot=True,
    cmap="coolwarm",
    fmt=".2f"
)

plt.title("Feature Correlation Heatmap")
plt.show()
```

### Exploratory Data Analysis Summary

The dataset contains 16,209 residential property records with 21 numerical and categorical features.
No missing values were observed in the dataset.

House prices show a right-skewed distribution, indicating the presence of high-value outliers.
Living area (sqft_living) exhibits a strong positive correlation with price, confirming that larger homes
tend to be more expensive. Latitude and longitude also show moderate correlation, highlighting the
importance of location in property valuation.

These observations motivate the use of both tabular features and satellite imagery to capture
structural as well as spatial characteristics of properties.
